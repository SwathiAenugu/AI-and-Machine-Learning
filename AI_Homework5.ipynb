**1.	Create random forest based on the following dataset in bootstrapping method taking the recommended number of subset selection (e.g. sqrt(n)) on the handouts as reference. And then write Python function to compare with your hand-analysis**

image.png

```python
import numpy as np
ds=np.array([[1,0,0,7,'Apple'],
             [0,1,0,20,'Watermelon'],
             [1,0,0,1,'Cherry'],
             [0,1,0,7.5,'Apple'],
             [1,0,0,1,'Strawberry'],
             [1,0,0,0.8,'Cherry']])
X=ds[:,:4]
y=ds[:,4]
print("X: \n",X)
print("\n Y: \n",y)
from sklearn.ensemble import RandomForestClassifier
rdmfr=RandomForestClassifier(bootstrap=True,random_state=5,max_features='sqrt',max_depth=2,n_estimators=100,criterion='gini')
rdmfr.fit(X,y)
tst=np.array([[0,1,1,0.4]])
print("\ntest:",tst)
rdmfr.predict(tst)
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
est=rdmfr.estimators_[5]
plt.figure()
plott=plot_tree(est,feature_names=['Red','Green','Blue','Size'],
                class_names=['Apples','Watermelon','Cherry','Strawberry'],
                filled=True,
                fontsize=10)
```

**2.	Given a function f(x)= e^(〖-x〗^2 )+0.01cos⁡(200x), find max f(x) value if x ∈ [-2, 2] in Python program by genetic algorithm, considering 1-digit precision of fractional decimal x. And then verify your program running result by the function plot curve in Python or Excel**

```python
import numpy as np
import matplotlib.pyplot as plt
import math
fx=[] 
x=np.arange(-2,2,0.02) #given x- range(-2,2)
for i in range(len(x)):
     x[i]=round(x[i],2)

print(x) 

for i in x:
    fx.append(math.e -i**2 + 0.01*(math.cos(200*i))) 
    
plt.xlabel("x ---->")
plt.ylabel('Y=f(x)')

plt.scatter(x,fx)
plt.plot(x,fx,'red',linewidth=2.5)
plt.show()
```
